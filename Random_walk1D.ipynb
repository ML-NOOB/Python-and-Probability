```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
prob = 0.5 #probability of moving forward 
```

```python
k = []
count = 0
for l in range (0,100):
 c = []
 s = 0
 for i in range(0,300):
     rand = random.uniform(0, 1) # we use a uniform distribution  varying iniformly between (0,1)
     if (rand < 0.5): # one step forward
        c.append(1)
     else:
        c.append(-1) # one step back
 s = sum(c)
 k.append(s)       
```

```python
c = []
s = 0
for i in range(0,300):
     rand = random.uniform(0, 1)
     if (rand > 0.5):
        c.append(1)
     else:
        c.append(-1)
for j in range(0,len(c)):
     s = s+c[j]
     plt.scatter(j,s, color = 'black')
plt.xlabel('step taken')
plt.ylabel('position')
plt.title('Random Walk')
```
