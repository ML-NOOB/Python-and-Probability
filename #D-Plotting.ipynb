```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python

fig1 = plt.figure( figsize=(10,6))
ax = plt.axes(projection='3d')
y,z = np.mgrid[-1:1:50j,-1:1:50j]
xline = y
yline = z

zline =(xline**2 + yline**2)
plt.figure(figsize=(8, 6))
ax.plot_wireframe(xline, yline, zline,color = 'g')
fig1.savefig('minima.jpg')
```

```python

fig2 = plt.figure( figsize=(10,6))
ax = plt.axes(projection='3d')
y,z = np.mgrid[-1:1:50j,-1:1:50j]
xline = y
yline = z

zline =-(xline**2 + yline**2)
plt.figure(figsize=(8, 6))
ax.plot_wireframe(xline, yline, zline,color = 'g')
fig2.savefig('maxima')
```

```python

fig3 = plt.figure( figsize=(10,6))
ax = plt.axes(projection='3d')
y,z = np.mgrid[-1:1:10j,-1:1:10j]
xline = y
yline = z

zline =(xline**2 - yline**2)
plt.figure(figsize=(8, 6))

ax.plot_wireframe(xline, yline, zline,color = 'g')


fig3.savefig('saddle')
```

```python
fig4 = plt.figure( figsize=(10,6))
xp = np.arange(-3, 4)
yp = np.arange(-3, 4)
zp  =np.ndarray((7,7))
for x in range(0, len(xp)):
    for y in range(0, len(yp)):
        zp[x][y] = -(xp[x]*xp[x] + yp[y]*yp[y])

plt.title('Contour plot for minima')
contours = plt.contour(xp, yp, zp)
plt.clabel(contours, inline=1, fontsize=12)
plt.show()
fig4.savefig('Cminima')
```

```python
fig5 = plt.figure( figsize=(10,6))
xp = np.arange(-3, 4)
yp = np.arange(-3, 4)
zp  =np.ndarray((7,7))
for x in range(0, len(xp)):
    for y in range(0, len(yp)):
        zp[x][y] = xp[x]*xp[x] + yp[y]*yp[y]

plt.title('Contour plot for maxima')
contours = plt.contour(xp, yp, zp)
plt.clabel(contours, inline=1, fontsize=12)
plt.show()
fig5.savefig('Cmaxima')
```

```python
fig6 = plt.figure( figsize=(10,6))
xp = np.arange(-3, 4)
yp = np.arange(-3, 4)
zp  =np.ndarray((7,7))
for x in range(0, len(xp)):
    for y in range(0, len(yp)):
        zp[x][y] = xp[x]*xp[x] - yp[y]*yp[y]

plt.title('Contour plot for saddle surface - Hyperbolic paraboloid')
contours = plt.contour(xp, yp, zp)
plt.clabel(contours, inline=1, fontsize=12)
plt.show()
fig6.savefig('Csaddle')
```

```python
fig7 = plt.figure( figsize=(6,4))
z =np.linspace(-np.pi,np.pi,100)
y =np.sin(z)
a,b= [1]*100,[-1]*100
plt.plot(z,y)
plt.plot(z,a)
plt.plot(z,b)
plt.title("Sine curve")
plt.xlabel('x')
plt.ylabel('sin(x)')
fig7.savefig('Sine_curve')
```
